```python
# Дана функция: f(x) = 5*x**2 + 10*x - 30
# 1. Определить корни +
# 2. найти интервалы, на которых функция возрастает
# 3. Найти интервалы, на которых функция убывает
# 4.Построить график +
# 5. Вычислить вершину +
# 6. определить промежутки, на которых f > 0
# 7. Определить промежутки, на которых f < 0
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a, b, c = 5, 10, -30
```

```python
def func(x):
    return a*x**2 + b*x + c
```

```python
limit = 5
step = 0.01
```

```python
x =np.arange(-limit, limit, step)
```

```python
def take_roots(a, b, c):
    discr = (b**2 - 4*a*c)
    if discr > 0:
        x1 = (-b + discr**0.5)/(2*a)
        x2 = (-b - discr**0.5)/(2*a)
        return round(x1, 2), round(x2, 2)
    elif discr == 0:
        x = -b/(2*a)
        return (round(x, 2),)
    else:
        return(None, )
```

```python
roots = take_roots(a, b, c)
print(roots)
```

```python
min_y = min(func(x))
min_x = take_roots(a,b,c - min_y)[0]
print(min_x, min_y)
```

```python
x_down_pos = np.arange(-limit, min(roots), step)
x_down_neg = np.arange(min(roots), min_x, step)
x_up_neg = np.arange(min_x, max(roots), step)
x_up_pos= np.arange(max(roots), limit, step)
```

```python
plt.rcParams['lines.linestyle'] = '-'
plt.plot(x_down_pos, func(x_down_pos), 'b', label = 'Убывание больше 0')
plt.plot(x_up_pos, func(x_up_pos), 'r', label = 'Возрастание больше 0')

plt.rcParams['lines.linestyle'] = '-.'
plt.plot(x_down_neg, func(x_down_neg), 'b', label = 'Убывание меньше 0')
plt.plot(x_up_neg, func(x_up_neg), 'r', label = 'Возрастание меньше 0')

plt.plot(roots[0], 0, 'rx', label = f'Корни ({roots[0]}, {roots[1]})')
plt.plot(roots[1], 0, 'rx')
plt.plot(min_x, min_y, 'go', label = f'Экстемум функции({min_x}, {min_y})')
plt.legend()
plt.grid()
```
